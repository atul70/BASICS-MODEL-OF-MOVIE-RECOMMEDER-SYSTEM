## NOW MAKE A RECOMMENDER SYSTEM FOR WHOLE DATASET..NOT ONLY ONE MOVIE

```python
userid_movietitle_matrix
```

```python
movie_corelation = userid_movietitle_matrix.corr(method= 'pearson', min_periods= 80)
movie_corelation
```

## NOW CREATE A DUMMY MOVIE FILE AND CHECK THIS DUMMY MOVIE RATINGS WITH THIS ABOVE RECOMMENDER SYSTEM

```python
myRatings = pd.read_csv('My_Ratings.csv')
myRatings
```

```python
similar_movie_list = pd.Series()
for i in range(0,2):
    similar_movie = movie_corelation[ myRatings['Movie Name'][i]].dropna()
    similar_movie = similar_movie.map( lambda x : x*myRatings['Ratings'][i])
    similar_movie_list = similar_movie_list.append(similar_movie)
```

```python
similar_movie_list.sort_values(inplace= True ,ascending= False)
similar_movie_list.head(50)
```
